```
from bertviz import head_view, model_view
from transformers import GPT2Tokenizer, GPT2Model
```

```
# Retreive attention weights

model_version = 'gpt2'
model = GPT2Model.from_pretrained(model_version, output_attentions=True)
tokenizer = GPT2Tokenizer.from_pretrained(model_version)

text = "The farmer offered apples to the housekeeper, because he had too many of them."
inputs = tokenizer.encode_plus(text, return_tensors='pt', add_special_tokens=True)
input_ids = inputs['input_ids']
attention = model(input_ids)[-1]
input_id_list = input_ids[0].tolist() # Batch index 0
tokens = tokenizer.convert_ids_to_tokens(input_id_list)
```

# Model View

The model view gives a birds-eye view of attention across all of the layers (rows) and heads (columns) in the model. In this case we are showing bert-base, which has 12 layers and 12 heads (zero-indexed).

```
from IPython.core.display import HTML
display(HTML('<script src="/static/components/requirejs/require.js"></script>'))
# Above two lines only needed when running in Colab
model_view(attention, tokens)
```

# Head View

The attention-head view visualizes attention in one or more heads in a particular layer in the model.
The head view allows us to observe the link between the pronoun and its antecedent

```
from IPython.core.display import HTML
display(HTML('<script src="/static/components/requirejs/require.js"></script>'))
# Above two lines only needed when running in Colab
head_view(attention, tokens)
```
